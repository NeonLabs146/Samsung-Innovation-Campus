###Implementation and Coding
###Algorithm to solve the coin exchange problem
####The input of the algorithm is the coin type and the amount of change, and the output is a list of coins. 
####At this time, suppose that the amount of coins is given in descending order.

```python
def coin_change(coins,amount):
  changes=[]
  largest=0
  while amount>0:
    if amount<coins[largest]:
      largest+=1
    else:
      changes.append(coins[largest])
      amount-=coins[largest]
  return changes
```

```python
coins=[500,100,50,10]
amount=int(input ('Input the amount'))
changes=coin_change(coins,amount)
print(changes,len(changes))
```

##3.1.**How the Triangle Path Finder works**
###A program that finds the sum of shortest paths in a triangle is a program that finds and outputs the sum of shortest paths in this triangle with the given list of numbers, as shown below.


```python
def find_minimum(row,col,triangle):
  if row==len(triangle):
    return 0
  else:
    minimum=min(find_minimum(row+1, col, triangle),
                find_minimum(row+1, col+1, triangle))
    return triangle[row][col]+minimum
```

```python
triangle=[
    [2],
    [3, 4],
    [6, 5, 7],
    [4, 1, 8, 3]
]
minimum=find_minimum(0,0,triangle)
print("The minimum sum is:",minimum)
```

#**Key concept**
##The Divide-and-Conquer Strategy
###The divide-and-conquer strategy divides the input instance of a given problem into two or more small input instances.
###If the answer to the divided input instance cannot be obtained immediately, divide it back into smaller input instances. 
###If you have obtained the answer to the divided input instance, you can integrate this answer to find the answer to the original input instance.
##**STEPS**
###**1.Divide:** Divide the input instance of the problem into two or more small input instances. 
###**2.Conquer:** Solve each of the divided input instances. If the divided instance is not small enough, use the recursion to solve it.
###**3.Combine:** If necessary, find the answer of the original input instance by combining the answers of the small input instance.

###bin_search() function from Unit 25
####This function used a strategy to reduce the scope of the search using low and high indices.

```python
def bin_search(nums,x):
  low,high = 0, len(nums)-1
  while low<=high:
    mid = (low+high)//2
    print(low,high,mid)
    if nums[mid]==x:
      return mid
    elif nums[mid]>x:
      high=mid - 1
    else:
      low=mid + 1
  return -1
```

##**1.2.Binary Search with the Divide-and-Conquer**
###The binary search problem can be implemented recursively, as shown below, by applying the divide-and-conquer strategy.


```python
def bin_search2(nums,x,low,high):
  if low>high:
    return -1
  else:
    mid = (low+high)//2
    print(low,high,nums[mid])
    if nums[mid]==x:
      return mid
    elif nums[mid]>x:
      return bin_search2(nums,x,low,mid-1)
    else:
      return bin_search2(nums,x,mid+1,high)
```

#**Key concept**
#**Dynamic Programming .vs. Divide-and-Conquer**
###Dynamic programming is the same method as the divide-and-conquer method, covered in Unit 31, in that both are recursive solutions to a given problem.
###Since the divide-and-conquer method repeats inefficient calculations when the overlapping subproblem occurs, the dynamic programming method stores the subproblems in a variable or table so that the already-calculated subproblems are not recalculated.
image.png

##**1.2The Design of Dynamic Programming Algorithms**
####• Dynamic programming is an algorithmic design technique that analyzes the recursive characteristics of a given problem and then obtains an answer in a solution in a bottom-up fashion. 
####At this time, the technique of storing the answers to the overlapping subproblems in the table in order to solve the overlapping subproblem is called memoization.
####----The development procedure of the dynamic programming algorithm is as follows:
#####i. Establish a recursive equation to calculate the answer to the input case of the problem.
#####ii. The answer to the entire input case is obtained by a bottom-up method that solves the small input case first.
#####iii. Dynamic programming uses memoization for bottom-up calculations.
#####iv. The answers for the subproblems are stored in a table (or memory).
#####v. When encountering the same subproblem, it is taken directly from the table (or memory) without recalculating.

##**1.2. n-th Fibonacci Term with Memoization**
####The fib2() function is implemented through a recursive function as a bottom-up solution by applying a dynamic programming method.


```python
F={0:0,1:1}
def fib2(n):
  if n in F:
    return F[n]
  else:
    F[n]=(fib2(n-1)+fib2(n-2))
    return F[n]
```

###The fib3() function implements memoization using a list instead of a dictionary.
###In this case, the value of the element of the Fibonacci term, which is not calculated, may be initialized to -1 to determine whether there is a duplicate calculation.


```python
def fib3(F,n):
  if n<=1:
    return F[n]
  else:
    if F[n]<0:
      F[n]=fib3(F,n-1)+fib3(F,n-2)
    return F[n]
```

##**1.3. Pascal’s Triangle by Recursion**
####Using the recursive properties of the binomial coefficient, bin(n, k) can be calculated as follows.


```python
def bin1(n,k):
  if k==0 or n==k:
    return 1
  else:
    return bin1(n-1,k-1) +bin1(n-1,k)
```

##**1.4. Pascal’s Triangle by Dynamic Programming**
####By applying dynamic programming, recursive expressions can be solved in a bottom-up manner using loops as shown below:


```python
def bin2(n,k):
  B=[[0]*(i+1) for i in range(n+1)]
  for i in range(n+1):
    for j in range(i+1):
      if j==0 or j==i:
        B[i][j]=1
      else:
        B[i][j]=B[i-1][j-1]+B[i-1][j]
  return B[n][k]
```

```python
def n_queens(i,col):
  if promissing(i,col):
    if i==len(col)-1:
      print(col)
    else:
      for j in range(len(col)):
        col[i+1]=j
        n_queens(i+1,col)
```

###Running the N-Queens Solver
###After receiving the value of N as an input from the user, the col list initializing the i-th row to -1 and the list elements to -1 is given as a parameter value, and the n_queens() function is called.

```python
N=int(input("Input the number of Queens:"))
n_queens(-1,[-1]*N)
```
