### Keyword Parameter

````python
def get_root(a, b, c):
    r1 = (-b + (b ** 2 - 4*a*c)**0.5) /(2*a)
    r2 = (-b - (b ** 2 - 4*a*c)**0.5) /(2*a)
    return r1, r2

result1, result2 = get_root(1, 2, -8)
print("the result is", result1, "or", result2)
````

image.png

- When calling a function, instead of passing only the value of the argument, the method of passing the argument by specifying the name is called a keyword parameter.
- In the code below, if -8, 2, 1 is given as an argument, the result is different from when 1, 2, -8 is given as an argument.


````python
result1, result2 = get_root(-8, 2, 1)
print("the result is", result1, "or", result2)
````

````python
# we can instead do this:

result1, result2 = get_root(a=1, b=2, c=-8)
print("the result is", result1, "or", result2)

result1, result2 = get_root(a=1, c=-8, b=2)
print("the result is", result1, "or", result2)

result1, result2 = get_root(b=2, c=-8, a=1)
print("the result is", result1, "or", result2)

# result of all 3 are same, the position of the argument is not important
````

````python
# positional argument however cannot folow keyword argument
result1, result2 = get_root(a=1, b=2, -8)
````

## Unit 18: Recursion Function Call

- A recursion is a function that calls itself inside a function. Recursion is a very useful troubleshooting technique. It can intuitively and simply solve problems that are difficult to solve with procedural techniques.
- Factorial expression and definition : n! = n * (n – 1)!, if n <= 1 : return 1


image.png

````python
def factorial(n):
    if n <= 1:
        return 1
    else:
        return n * factorial(n-1)
n = 5 
print('{}! = {}'.format(n, factorial(n)))
````

````python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(10))
````

````python
%%timeit -n 1 -r 20
fibonacci(30)
````

### Closure

- The basic form of a closure function is as follows. It creates an nested function inside a function and returns the nested function as a return value.
- Here you can say that the nested function ‘was closured.’ If the function is closured, the variable inside the body of the original function (outer function) does not disappear from the memory. Then, in the next call it can utilize the nested function.
- Study the concept of the closure with the following code. 
- What are free variables? A free variable is only used in a certain code block, but it is not a global variable, and it is not defined within the block. It is a relative concept. 
- That is, in the point of view of mult function, a is a free variable. However, for closure-calc, a is only a local variable. 


````python
def closure_calc():
    a = 2
    def mult(x):
        return a * x
    return mult

c = closure_calc()
print(c(1), c(2), c(3))
````

#### Usage of Closure

1. Limiting usage of global variables 
2. Data hiding 


- A closure has a separate memory for each variable. 


image.png


````python
def makecounter():
    count = 0

    def counter():
        nonlocal count
        count += 1
        return count
    return counter

c1 = makecounter()
c2 = makecounter()
print("c1", c1())
print("c1", c1())
print("c2", c2())
````

### Change Local Variable of Closure


*   Use the nonlocal to change the local variable of a closure



````python
def calc():
    a = 2
    b = 3
    total = 0
    def mult_add(x):
        nonlocal total
        total = total + a * x + b
        return total
    return mult_add
c = calc()
print(c(1), c(2), c(3))
````

#### Searcing with regular expression:

- In case of a text string as follows, the compile function of the re module searches regular expression patterns and compiles them to regular expression objects.
- Then, a regular expression object called regex is created, and this object searches text with a given pattern. 
- \d{3} finds patterns of 3 integers. \d{4} finds patterns of 4 integers. The bracket places these patterns into group 1 and group 2.


````python
import re

text = 'please call 010-2345-1234'
# has two groups (\d{3}) and (\d{4}-\d{4})
regex = re.compile('(\d{3})-(\d{4}-\d{4})')
match_obj = regex.search(text)

# the first one is the whole search
print(match_obj.group())
````

````python
print(match_obj.group(1))
print(match_obj.group(2))
````

### Class and Inheritance

Inheritance in python can be done as follows where class B is inheriting from class A

```
class A:
    ...

class B(A):
    ...

```

**Class of hierarchical structure:**
- Examine the class with hierarchical structure below. 
- You can construct class Manager that means manager and class Employee that means employee by inheriting the class Person, which is the parent class. 
- Since managers and employees are both human-beings, the two child classes should contain the name attribute and get_name() method from the Person class. 
- If the class Manager and the class Employee inherit the class Person, the two classes will contain the member variables and member functions of the class Person without any extra coding.



image.png

````python
class Person:
    def __init__(self, name):
        self.name = name
    def get_name(self):
        return self.name

class Employee(Person):
    """
    Child class of person class
    """
    def __init__(self, name, staff_id):
        Person.__init__(self, name)
        self.staff_id = staff_id

    def info(self):
        return "staff: " + self.get_name() + ", staffID:"+ str(self.staff_id)

class Manager(Person):
    """
    Child class of person class
    """
    def __init__(self, name, position):
        Person.__init__(self, name)
        self.position = position

    def info(self):
        return "Manager: " + self.get_name() + ", position:"+ str(self.position)

worker1 = Employee("David Doe", 1111)
worker2 = Employee("Paul Carter", 2222)
cfo = Manager("Anna Smith", "CFO")
print(worker1.info())
print(worker2.info())
print(cfo.info())
````
