#### PP 22
Agency A is planning to issue tickets of a concert hall for idol singers' concert. Here, the number n is the input and the seat number is arranged as follows. n * n seats are placed when n is given as input. The below arrangement of the seat numbers is called a snake matrix because the array increases in ones in the shape of a snakeâ€™s trunk. Write a program that produces arrays of these numbers.
###image.png

```python
numRows = 5
num = 1;

for rowVal in range(1, numRows + 1):
    for colVal in range(1, numRows + 1):
        if rowVal % 2 == 0:                                   #even row
            print("%3d"%(num), end=" ")
            num -= 1
        else:                                                 #odd row
            print("%3d"%(num), end=" ")
            num += 1
    num += numRows + (-1 if rowVal % 2 == 1 else 1);   # this is how we can use ternary operator in python
    print()   # for new line
```

```python
#@title
def generate_snake_matrix(n):
    matrix = [[None] * n for _ in range(n)]
    row, col = 0, 0
    direction = "right"
    for i in range(1, n**2+1):
        matrix[row][col] = i
        if direction == "right":
            if col == n-1 or matrix[row][col+1] is not None:
                direction = "down"
                row += 1
            else:
                col += 1
        elif direction == "down":
            if row == n-1 or matrix[row+1][col] is not None:
                direction = "left"
                col -= 1
            else:
                row += 1
        elif direction == "left":
            if col == 0 or matrix[row][col-1] is not None:
                direction = "up"
                row -= 1
            else:
                col -= 1
        elif direction == "up":
            if row == 0 or matrix[row-1][col] is not None:
                direction = "right"
                col += 1
            else:
                row -= 1
    return matrix


n=int(input("Enter the size of Matrix:"))
mat=generate_snake_matrix(n)
for row in mat:
    print(" ".join(str(x) for x in row))
```

##**Number Guessing Game**

```python
import random

Jackpot = random.randint(1,100)

guess = int(input("Guess the number: "))
count = 1

while guess != Jackpot:
  if guess < Jackpot:
    print("Wrong Guess!!!, Guess higher")
  else:
    print("Wrong Guess!!, Guess lower")

  guess = int(input("Guess the number: "))
  count+=1
else:
  print("Correct Guess!!!")
  print("No of attempts:", count)

```
