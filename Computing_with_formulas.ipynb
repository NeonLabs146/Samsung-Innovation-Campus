 Compute s in meters when s = v0t + 1/2(at^2), with v0 = 3 m/s, t = 1 s, a =
2 m/s^2.

```python
v0=3
t = 1
a = 2
s=v0*t + (1/2)*a*t**2
print(s)
```

```python
import json
import csv

with open('/content/GFCT_USGS_Station .json', 'r') as json_file:
    data = json.load(json_file)

csv_file_name = 'GFCT_USGS_Station.csv'

# Write CSV data
with open(csv_file_name, 'w', newline='') as csv_file:
    # Create a CSV writer
    csv_writer = csv.writer(csv_file)

    if data:  # Check if data is not empty
        # Write header
        headers = data['Field Data'][0].keys()
        csv_writer.writerow(headers)


        # Write data
        for entry in data['Field Data']:
          csv_writer.writerow(entry.values())
    else:
        print("No data to convert.")

print(f'Conversion complete. CSV file saved as {csv_file_name}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read CSV file into a DataFrame
df = pd.read_csv('GFCT_USGS_Station.csv')

# Plotting with seaborn and matplotlib
sns.set(style="whitegrid")
df['Time'] = pd.to_datetime(df['Time'], format='%Y:%m:%d %H:%M:%S.%f +0000')

# Example plot: EXt vs Time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Time', y='EXt', data=df, marker='o', label='EXt')
plt.title('EXt over Time')
plt.xlabel('Time')
plt.ylabel('EXt')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
